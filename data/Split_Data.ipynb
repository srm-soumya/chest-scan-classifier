```python
import pandas as pd
import os
import shutil
```

```python
data = pd.read_csv('clean_data.csv')
```

```python
mapper = data[['Image Index', 'label']]
```

```python
d = mapper.set_index('Image Index').to_dict()['label']
```

```python
d
```

```python
sets = 'sets'
os.makedirs(os.path.join('sets'))
```

```python
for file in os.listdir('sample'):
    try:
        dir = d[file]
        if not os.path.exists(os.path.join(sets, dir)):
            os.makedirs(os.path.join(sets, dir))
        shutil.copy(os.path.join('sample', file), os.path.join(sets, dir, file))
    except:
        pass
```

```python
from random import shuffle
cwd = os.getcwd()

for d in os.listdir(sets):
    os.makedirs(os.path.join(cwd, 'train', d))
    os.makedirs(os.path.join(cwd, 'validation', d))
    os.makedirs(os.path.join(cwd, 'test', d))
    
    files = os.listdir(os.path.join(cwd, sets, d))
    shuffle
    size = len(files)
    split1 = int(0.7 * size)
    split2 = int(0.85 * size)
    
    for f in files[:split1]:
        src = os.path.join(cwd, sets, d, f)
        dest = os.path.join(cwd, 'train', d, f)
        shutil.copy(src, dest)
        
    for f in files[split1:split2]:
        src = os.path.join(cwd, sets, d, f)
        dest = os.path.join(cwd, 'validation', d, f)
        shutil.copy(src, dest)
        
    for f in files[split2:]:
        src = os.path.join(cwd, sets, d, f)
        dest = os.path.join(cwd, 'test', d, f)
        shutil.copy(src, dest)
```
